```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Diameter_inch":[8,10,12],
    "Price_dollar":[10,13,16]
}
df = pd.DataFrame(data)
df
```

```python
X = np.array(df["Diameter_inch"])
Y = np.array(df["Price_dollar"])
```

```python
mean_X = X.mean()
mean_Y = Y.mean()
```

```python
numerator = np.sum((X - mean_X) * (Y - mean_Y))
denominator = np.sum((X - mean_X) ** 2)
m = numerator / denominator
m # slope  = 1.5
# calculate intercept(b)
# b = Y-mX 
b= mean_Y -( m * mean_X)
b
```

```python
#calculate regresion, regression(Y) = mX - b
def predict_price(diameter):
    return m * diameter + b
predict_price(20)
```

```python
# visualization(Data + Regression line)
plt.figure()
```

```python
plt.scatter(X,Y)
plt.plot(X,m*X + b)
plt.xlabel("Pizza Diameter(inches)")
plt.ylabel("Pizza Price (dollars)")
plt.title("Pizza Price prediction usinng Linear Regression")
plt.show()
```

```python
# Visualization with prediction point
plt.figure()
plt.scatter(X,Y)
plt.plot(X,m*X+b)
plt.scatter(12,predict_price(12))
plt.xlabel("Pizza Diameter(inches)")
plt.ylabel("Pizza price (dollars)")
plt.title("20-inch Pizza prediction")
plt.show()
```
